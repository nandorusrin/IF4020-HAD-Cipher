```python
import feistelChiperWithFunction as fc
import feistelFunction as ff
import sampleText
```

```python
print("Fiestel CHIPER TESTING")

function_class = ff.FeistelFunction()

message = bytearray(sampleText.pride_and_prejudice,"raw_unicode_escape")
cipherMachine = fc.FeistelNetwork(key=bytearray('csdfghjkrwertyuj',"raw_unicode_escape"),f_function = function_class )
    
    
print("The plain text is: ")
# print(message.decode("raw_unicode_escape"))
print()

ciphertext = cipherMachine.encrypt_message(message)
print("The encrypted message is: ")
# print(ciphertext.decode('raw_unicode_escape'))
print("==========================================================")
plaintext = cipherMachine.decrypt_cipher(ciphertext)
# print("The decrypted message is: ")
print(plaintext)
print()
    
```

```python
freq_dct_msg= {}

for a in message:
    freq_dct_msg[a] = freq_dct_msg.get(a,0) + 1
```

```python
freq_msg_decrypted = {}
for a in plaintext:
    freq_msg_decrypted[a] = freq_msg_decrypted.get(a,0) + 1
```
